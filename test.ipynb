```python
# Rename all column labels to replace spaces with underscores and convert everything to lowercase. (Underscores can be much easier to work with in Python than spaces.)
def column_name(df1):
    df2 = df1.copy()
    df2 = df2.rename(columns=lambda x: x.strip().lower().replace(" ", "_"))
    return df2
```

```python
def dropRedundant(df1):
    df2 = df1.copy()
    df3 = df2.drop(['vehicle_id', 'engine_index'], axis=1)
    df4 = df3.drop(['transmission_descriptor'], axis=1)
    df_final = df4.drop(['alternative_fuel/technology', 'electric_motor', 'manufacturer_code', 'gasoline/electricity_blended_(cd)',
                   'vehicle_charger', 'alternate_charger', 'hours_to_charge_(120v)', 'hours_to_charge_(240v)', 
                    'hours_to_charge_(ac_240v)', 'range_(ft1)', 'city_range_(ft1)', 'highway_range_(ft1)',
                   'range_(ft2)', 'city_range_(ft2)', 'highway_range_(ft2)', 'gas_guzzler_tax', 'save_or_spend_(5_year)'], axis=1)
    
    return df_final
```

```python
def fueltype(df1):
    df2 = df1.copy()
    df3 = df2[df2['fuel_type_2'].isna()]
    df4 = df3.drop(['fuel_type', 'fuel_type_2', 'city_mpg_(ft2)', 'highway_mpg_(ft2)', 'unadjusted_city_mpg_(ft2)',
                   'unadjusted_highway_mpg_(ft2)', 'combined_mpg_(ft2)', 'annual_fuel_cost_(ft2)', 
                   'annual_consumption_in_barrels_(ft2)', 'tailpipe_co2_(ft2)', 'tailpipe_co2_in_grams/mile_(ft2)',
                   'composite_city_mpg', 'composite_highway_mpg', 'composite_combined_mpg'], axis=1)
    return df4
```

```python
def cleanDataset(data):
    df1 = data.copy()
    df1 = column_name(df1)
    df1 = unro(df1)
    df1['drive'] = cleanDrive(df1['drive'])
    df1['start_stop_technology'] = cleanStart(df1['start_stop_technology'])
    df1['turbocharger'] = cleanCharger(df1['turbocharger'])
    df1['supercharger'] = cleanCharger(df1['supercharger'])
    df1 = dropNull(df1)
    df1 = dropRedundant(df1)
    return df1
```

# Did fuel consume change radically after millenium?

```python
year2000 = data1[data1['year'] >= 2000].groupby('class')['combined_lkm'].mean()
```

```python
year2000u = data1[data1['year'] < 2000].groupby('class')['combined_lkm'].mean()
```

```python
year2000u
```

```python
diff = year2000u- year2000
```

```python
diff.dropna(inplace=True)
plt.subplots(figsize=(8, 5))
plt.bar(diff.index,diff)
plt.title('Improvements in Fuel Economy after millenium by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('Reduction after 2000')
plt.xticks(rotation=90);
```
