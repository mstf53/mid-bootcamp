```python
def dropRedundant(df1):
    df2 = df1.copy()
    df3 = df2.drop(['vehicle_id', 'engine_index'], axis=1)
    df4 = df3.drop(['transmission_descriptor'], axis=1)
    df_final = df4.drop(['alternative_fuel/technology', 'electric_motor', 'manufacturer_code', 'gasoline/electricity_blended_(cd)',
                   'vehicle_charger', 'alternate_charger', 'hours_to_charge_(120v)', 'hours_to_charge_(240v)', 
                    'hours_to_charge_(ac_240v)', 'range_(ft1)', 'city_range_(ft1)', 'highway_range_(ft1)',
                   'range_(ft2)', 'city_range_(ft2)', 'highway_range_(ft2)', 'gas_guzzler_tax', 'save_or_spend_(5_year)'], axis=1)
    
    return df_final
```

```python
#engine_displacement_median =  dict(data.groupby('engine_displacement')['combined_lkm'].agg('median').sort_values())

def map_values(df, col, col_to_aggregate):
    df2=df.copy()
    col_aggregated = df2.groupby(col)[col_to_aggregate].agg("median").sort_values()
    col_median = dict(col_aggregated)
    col_median_dict = {key: value/col_median[col_aggregated.index[0]] for key, value in col_median.items()}
    df2[col].replace(col_median_dict, inplace=True)

    return df2
```

```python
data = map_values(data, "engine_displacement", "combined_lkm")
```

```python
data = map_values(data, "engine_cylinders", "combined_lkm")
```
